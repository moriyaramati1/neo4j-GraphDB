Queries to understand the data-

Viewing the number of relationships created in the graph:


````python
# connected to the neo4j graph we have created, change according to your personal path. 
g = GraphDatabase.driver("bolt://localhost:7687", auth=("neo4j", "1234"))

# we would like to get a sense of the data and understand how many relationships we have created. 
# therefore we will count how many times each relationships appears. 
result = {"relType": [], "count": []} # at the begining it is empty, we will count how many realtionships we have from each type. 
with g.session() as session: 
    # checking what type of relationships we have and adding to a new list. 
    rel_types = [row["relationshipType"] for row in session.run("CALL db.relationshipTypes()")]
    for rel_type in rel_types: # for each relationship see how many we have. 
        # match any type of node that contains the relationship we are running on and is connected to another node.
        # return the number of times it appears. 
        query = f"MATCH ()-[:`{rel_type}`]->() RETURN count(*) as count"
        count = session.run(query).single()["count"] # adding the count to a new list. 
        result["relType"].append(rel_type) # adding the types of relationships to result.
        result["count"].append(count) # adding the number of relationships to result. 

relation = pd.DataFrame(data=result) # convert to a dataframe. 
print(tabulate(relation.sort_values("count"), headers='keys',tablefmt='psql', showindex=False)) # printing the results.
print(type(relation)) # viewing relations type. 
relation # viewing relation. 


````

building the clinic similarity relationships and graph-

````python
# if it is your first time runing this code you must run the code in the comment in neo4j sandbox
# (the notebook had problems runing it from here, and dont running it more then once since it was already created)
'''
# create a sub graph called 'Similarity_Graph_Clinic' which contains Person and Gene nodes and a relationship Has_a. 
query ="CALL gds.graph.create('Similarity_Graph_Clinic',['Person', 'Gene'],'Has_a')"
graph.run(query) # run the query. 
'''
# we would like to add relationships on the 'Similarity_Graph_Clinic' graph, the relationship is called Clinic_Similarity and it has a 
# property 'SIMILARITY' which will contain the level of the two nodes similarity according to the algorithm. 
query="CALL gds.nodeSimilarity.write('Similarity_Graph_Clinic',{writeRelationshipType: 'Clinic_Similarity',writeProperty: 'SIMILARITY'})"
graph.run(query) # run the query. 
# creating the similarity and returning every two person nodes and their value of similarity in a descinding order. 
query="CALL gds.nodeSimilarity.stream('Similarity_Graph_Clinic')YIELD node1, node2, similarity RETURN gds.util.asNode(node1).ID AS Person1, gds.util.asNode(node2).ID AS Person2, similarity ORDER BY similarity DESCENDING, Person1, Person2"
df7=graph.run(query).to_data_frame() # create a dataframe from the query.

````

if any of the inputs arent correct a message will appear.

````python

# ask the user what type of similarity he would like. 
print('Which type of similarity do you want(based on clinic or not clinic):')
inp= input() # recieve answer. 
flag=0 # helper. 

while(flag==0): # while flag is 0 we didnt recieve a proper answer from the user. 
    # clinic similarity. 
    if (inp=='clinic'):
        # ask for the patients ID.
        print('Type the ID of person ')
        name=input() # recieve answer. 
        print() # space. 
        flag=1 # change flag. 
        # try to match a person node to the id recieved, if the id dosnt exist.
        while( not nodes.match("Person", ID=name).first()):  
            print("No such patient exists, please try again") # ask to recieve a new id. 
            name=input() # recieve answer. 
            print() # space. 
        
        # recieve all the people that share a similarity with the node, according to df7. 
  
        x=df7[df7["Person1"]==name] # Takes a sub-data frame that will contain all the lines which the patient we want appears in. And patients similar to him.
        row_1=x.iloc[0]   #get the first and most similar pateint
        sim=row_1['similarity'] # Access to the first patient similarity score because he has the highest level of similarity  
        #There may be several patients with the same maximum level of similarity. A sub-data frame was built that will contain all patients who have this specific level of similarity to the requested patient.
        y=x[x['similarity']==sim] 
        print("the similarity is:" ,sim) # Print the highest similarity.
        print() # print space

        #We want to know why those patients similar to the requested patient, for that reason we check each similar patient which clinic attribute they share.
        for i in range(0,len(y)): # For each simialr patient: 
            person=y['Person2'].iloc[i] # Gets the ID of this patient
            print(person) # Print his ID
            
            # we built a specific query with this patient ID and the requested patient, and check which gene they share.
            query="match(p1:Person{ID:'"+person+"'})-->(g:Gene)<--(p2:Person{ID:'"+name+"'}) return g" 
            #print(query)
            
            # we run this query, and put the result into data frame.
            d=graph.run(query).to_data_frame()
            # print the data frame with the gene they share.
            print(d)

    # not clinic similarity.      
    elif(inp=='not clinic'):
        # ask for the patients ID.
        print('Type the name of person ')
        name=input() # recieve answer. 
        print() # space. 
        flag=1 # change flag. 
        # try to match a person node to the id recieved, if the id dosnt exist.
        while( not nodes.match("Person", ID=name).first()):
            print("No such patient exists, please try again")
            name=input() # recieve anwswer.
            print() # space. 
        
        # recieve all the people that share a similarity with the node, according to df8. 
        x=df8[df8["Person1"]==name] # Takes a sub-data frame that will contain all the lines which the patient we want appears in. And patients similar to him.
        row_1=x.iloc[0]    #get the first and most similar pateint
        sim=row_1['similarity'] # Access to the first patient similarity score because he has the highest level of similarity 
        
        #There may be several patients with the same maximum level of similarity. A sub-data frame was built that will contain all patients who have this specific level of similarity to the requested patient.
        y=x[x['similarity']==sim] 
        print("the similarity is:" ,sim) # Print the highest similarity.
        print() # print space
        
        # We want to know why those patients similar to the requested patient, for that reason we check each similar patient which non clinic attribute they share.

        for i in range(0,len(y)): # For each simialr patient:

            person=y['Person2'].iloc[i] # Gets the ID of this patient
            print(person)  # Print his ID
            
            # we built a specific query with this patient ID and the requested patient, and check which gene they share.
            query="match(p1:Person{ID:'"+person+"'})-->(g)<--(p2:Person{ID:'"+name+"'}) where not(g:Person) and not(g:Gene) return g"
            #print(query)
            # we run this query, and put the result into data frame.
            d=graph.run(query).to_data_frame()
            # print the data frame with the gene they share.
            print(d)
                 
    else:
      # If the user enters something different from clinical or non-clinical, an error is returned to him, and ask him to try again
        flag=0
        print('Which type of similarity do you want(based on clinic or not clinic):')
        inp= input()
````
